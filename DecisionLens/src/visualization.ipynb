## Visualization part 

- This modules aims to get the undersatding on data and available relations if present using visual entities like graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'../data/raw/simulated_data.csv')
```

```python
data.head()
```

#### Distribution of data

```python
fig, ax = plt.subplots(
    nrows=2,
    ncols=2,
    figsize = (12,8)
)

ax = ax.flatten()
sns.histplot(data=data, x="age", kde=True, ax=ax[0])
sns.histplot(data=data, x="session_duration", kde=True, ax=ax[1])
sns.histplot(data=data, x="previous_orders", kde=True, ax=ax[2])
sns.histplot(data=data, x="purchase_prob", kde=True, ax=ax[3])

for axis in ax:
    for container in axis.containers:
        axis.bar_label(container, fmt="%d", fontsize=8)

plt.tight_layout()
```
